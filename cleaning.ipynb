```python
import pandas as pd
import numpy as np
```

```python
redemption_data = pd.read_csv("data/eb_redemption_details.csv")
redemption_data.columns = ["sl_no", "date_of_encashment", "party_name", "party_acc_no", "prefix", "bond_no", "denom", "pay_branch_code", "pay_teller"]
```

```python
redemption_data.head()
```

```python
                                  
redemption_data = redemption_data[redemption_data['party_name'] != 'Name of the Political Party']
redemption_data['party_name'].unique()
```

```python
purchase_data = pd.read_csv("data/eb_purchase_details.csv")
purchase_data.columns = ['sl_no', 'urn', 'journal_date', 'purchase_date', 'expiry_date', 'purchaser', 'prefix', 'bond_no', 'denom', 'issue_branch_code', 'issue_teller'
                         , 'status']
```

```python
purchase_data = purchase_data[purchase_data['purchaser'] != "Name of the Purchaser"]
purchase_data.head()
```

```python
redemption_data.to_csv("data/eb_redemption_details.csv", index=False)
purchase_data.to_csv("data/eb_purchase_details.csv", index=False)
```
